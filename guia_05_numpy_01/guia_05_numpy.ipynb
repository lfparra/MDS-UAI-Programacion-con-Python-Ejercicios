```python
notas = np.random.random([40,3])*6+1 #matriz con notas aleatorias
notas
```

```python
prom = np.zeros([40]) #arreglo para guardar los promedios
prom
```

```python
for i in range(0, notas.shape[0]):
  prom[i] = notas[i,:].mean()
  print(prom[i])
```

```python
print(f'Los promedios son {prom.round(2)}')
```

**Ejercicio 02**

```python
arreglo_02 = np.random.random([100])
arreglo_02
```

```python
promedio_inicial = np.mean(arreglo_02)
promedio_inicial
```

```python
nuevo_arreglo = arreglo_02 - promedio_inicial
nuevo_arreglo
```

```python
print(f'Los número menores a 0 son {np.sum(nuevo_arreglo < 0)}')
print(f'Los número mayores a 0 son {np.sum(nuevo_arreglo > 0)}')
```
